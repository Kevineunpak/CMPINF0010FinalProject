```python
import matplotlib.pyplot as plt
import pandas as pd
# Load the PGH 311 Complaint data
data_url = \
"https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f"
complaints = pd.read_csv(data_url)
# Produce a graph that shows the number of complaints per month
complaints.head()
```

```python
c = complaints["INCIDENTNEIGHBORHOOD"].value_counts().to_frame().sort_values(by="count", ascending=True)
c.plot.bar(title="Number of Arrests per Neighborhood", figsize=(20,8))
```

From the graph, we can see that the neighborhoods with the least number of arrests are
1. Mt. Oliver Neighborhood :    2
2. Troy Hill-Herrs Island  :    6
3. Mt. Oliver Boro         :   18
4. Central Northside       :   23
5. Regent Square           :   35
