```python
%matplotlib inline
import matplotlib.pyplot
import numpy as np
import pandas as pd

data_url = \
"https://data.wprdc.org/datastore/dump/632fbb91-c55d-4221-a8ad-91c72902bc61"

rawData = pd.read_csv(data_url)
rawData.head(5)
```

```python
clean = rawData["neighborhood"]
clean.to_frame()
```

```python
graph = rawData["neighborhood"].value_counts().to_frame().sort_values(by="neighborhood", ascending=True)

graph.plot.bar(title="Number of Crosswalks", figsize=(20,8))
```
